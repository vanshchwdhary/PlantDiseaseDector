```python
import pandas as pd
import pickle

def predict_fertilizer(temparature, humidity, moisture, nitrogen, potassium, phosphorous, soil_type, crop_type):

    # Load model and label encoder
    with open("fertikizer.pkl", "rb") as f:
        model = pickle.load(f)
    with open("label_encoder.pkl", "rb") as f:
        le = pickle.load(f)

    # Define all possible columns as per training data
    columns = [
        'Temparature', 'Humidity', 'Moisture', 'Nitrogen', 'Potassium', 'Phosphorous',
        'Soil Type_Black', 'Soil Type_Clayey', 'Soil Type_Loamy', 'Soil Type_Red', 'Soil Type_Sandy',
        'Crop Type_Barley', 'Crop Type_Cotton', 'Crop Type_Ground Nuts', 'Crop Type_Maize',
        'Crop Type_Millets', 'Crop Type_Oil seeds', 'Crop Type_Paddy', 'Crop Type_Pulses',
        'Crop Type_Sugarcane', 'Crop Type_Tobacco', 'Crop Type_Wheat'
    ]

    # Prepare input dictionary
    input_dict = {col: [0] for col in columns}
    input_dict['Temparature'] = [temparature]
    input_dict['Humidity'] = [humidity]
    input_dict['Moisture'] = [moisture]
    input_dict['Nitrogen'] = [nitrogen]
    input_dict['Potassium'] = [potassium]
    input_dict['Phosphorous'] = [phosphorous]
    input_dict[f'Soil Type_{soil_type}'] = [True]
    input_dict[f'Crop Type_{crop_type}'] = [True]

    input_df = pd.DataFrame(input_dict)
    pred = model.predict(input_df)[0]
    return le.inverse_transform([pred])[0]
```

```python
prediction = predict_fertilizer(26, 52, 38, 37, 0, 0, 'Sandy', 'Maize')
print("Predicted Fertilizer:", prediction)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pickle

# Load data
data = pd.read_csv("Fertilizer Prediction.csv")

# Target column
target_col = 'Fertilizer Name'

# Separate features and target
X = data.drop(target_col, axis=1)
y = data[target_col]

# One-hot encode categorical features
X = pd.get_dummies(X)

# Label encode target
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Save label encoder
with open("label_encoder.pkl", "wb") as f:
    pickle.dump(le, f)

# Stratified train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, stratify=y_encoded, random_state=42)

# Try RandomForest with more estimators
model = RandomForestClassifier(n_estimators=300, max_depth=15, min_samples_split=4, random_state=42)

# Cross-validation (more robust)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='accuracy')

print(f"📊 Cross-validation Accuracy: {cv_scores.mean() * 100:.2f}%")

# Train final model on full training data
model.fit(X_train, y_train)

# Evaluate on test data
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"✅ Final Test Accuracy: {accuracy * 100:.2f}%")

# Save model
with open("ferti_try.pkl", "wb") as f:
    pickle.dump(model, f)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Train Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Evaluate
dt_pred = dt_model.predict(X_test)
dt_acc = accuracy_score(y_test, dt_pred)
print(f"🌳 Decision Tree Test Accuracy: {dt_acc * 100:.2f}%")

# Save the model
with open("Navis_Ferti_DT.pkl", "wb") as f:
    pickle.dump(dt_model, f)
```
